```python
# 参照:2013年9月20日名古屋競馬1レース

tansyo = np.array([14.7, 72.6, 1.4, 2.3, 151.8, 18.0, 66.8])
umatan1 = np.array([np.nan, 347.2, 37.4, 54.0, 662.7, 169.6, 607.5])
umatan2 = np.array([455.6, np.nan, 177.8, 197.1, 1457.9, 331.4, 2429.8])
umatan3 = np.array([23.0, 86.8, np.nan, 5.4, 50.7, 13.0, 177.8])
umatan4 = np.array([10.3, 49.3, 1.9, np.nan, 90.0, 5.9, 123.6])
umatan5 = np.array([1214.9, 1457.9, 74.4, 455.6, np.nan, 560.8, 3644.7])
umatan6 = np.array([119.5, 347.2, 27.6, 43.2, 486.0, np.nan, 455.6])
umatan7 = np.array([1041.4, 2429.8, 1214.9, 911.2, 3644.7, 911.2, np.nan])
umatan = [umatan1, umatan2, umatan3, umatan4, umatan5, umatan6, umatan7]

dic = {'単勝':tansyo}
for i in range(len(tansyo)):
    dic[str(i+1)] = umatan[i]

df = pd.DataFrame(dic, index = [str(i+1) for i in range(len(tansyo))])
df
```

```python
re_df = Relative(df).sum_odds()
re_df
```

```python
Choise(re_df).combine()
```
